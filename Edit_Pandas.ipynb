3) Convert all latitude and longitude values to decimal degrees. You may need to check and correct formatting of Latitude and Longitude values in the process. The formula to convert from degrees minutes seconds is: decimal_degree = degrees + min/60 +sec/3600. The formula to convert from decimal minutes is: decimal_degree = degrees + min/60

```python
import arcpy

line=0
for index, row in newdata.iterrows():
    line+=1
    if row["Latitude Units"]=='deg/min/sec': 
        lat=row["Latitude"].split("/")
        d=float(lat[0])
        m=float(lat[1])
        s=float(lat[2])
        latdec=(d+(m/60)+(s/3600))
    elif row["Latitude Units"]=='decimal minutes':
        lat=row["Latitude"].split("/")
        d=float(lat[0])
        m=float(lat[1])
        latdec=(d+(m/60))
    elif row["Latitude Units"]=='decimal degrees': 
        latdec=row["Latitude"]
    if row["Longitude Units"]=='deg/min/sec':
        lon=row["Longitude"].split("/")
        d=float(lon[0].replace('\U00002013', '-'))
        m=float(lon[1])
        s=float(lon[2])
        londec=(d-(m/60)-(s/3600))
        if londec>0:
            londec=-londec
    elif row["Longitude Units"]=='decimal minutes':
        lon=row["Longitude"].split("/")
        d=float(lon[0].replace('\U00002013', '-'))
        m=float(lon[1])
        londec=(d-(m/60))
        if londec>0:
            londec=-londec
    elif row["Longitude Units"]=='decimal degrees':
        lon=row["Longitude"]
        if isinstance(lon, float):
            londec=lon
        else:
            londec=float(lon.replace('\U00002013', '-'))
        
    print (latdec, londec)

print (f" The total number of points is: {line}")
```
